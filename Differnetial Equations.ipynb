### Solving differential equations in the python

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#function that returns dy/dt
def model(y,t,k):
    dydt = -k*y
    return dydt
# initital condition
y0 = 5
#time points
t = np.linspace(0,20)
# solve ODE
k = 0.1
y1 = odeint(model,y0,t, args = (k,))
k = 0.2
y2 = odeint(model,y0,t, args = (k,))
k = 0.5
y3 = odeint(model,y0,t, args = (k,))

# plot results
plt.plot(t,y1, 'r-' , linewidth=2, label='k=0.1')
plt.plot(t,y2, 'b--' , linewidth=2, label='k=0.2')
plt.plot(t,y3, 'g:' , linewidth=2, label='k=0.5')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

### Problem 1

$\frac{dy}{dt} = -y(t) + 1$ with y(0) = 0

```python
def problem1(y,t):
    dydt = -y+1.0
    return dydt
#initial condition
y0 = 0
#time points
t = np.linspace(0,5)
#solve ODE
y = odeint(problem1,y0,t)
#plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show
```

```python
##hiv simulation
def hiv(x,t):
    kr1 = 1e5
    kr2 = 0.1
    kr3 = 2e-7
    kr4 = 0.5
    kr5 = 5
    kr6 = 100
    h = x[0]
    i = x[1]
    v = x[2]
    p = kr3*h*v
    dhdt = kr1 - kr2*h - p
    didt = p - kr4*i
    dvdt = -p -kr5*v + kr6*i
    return [dhdt, didt, dvdt]
x0 = [1e6,0,100]
t = np.linspace(0,15,1000)
x = odeint(hiv,x0,t)

h = x[:,0]
i = x[:,1]
v = x[:,2]
plt.semilogy(t,h)
plt.semilogy(t,i)
plt.semilogy(t,v)
plt.show()
```
